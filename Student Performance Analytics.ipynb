```python
 # Extracting categorical and numerical features
categorical_features = df.select_dtypes(include=['object','category']).columns.tolist()
numerical_features = df.select_dtypes(include=['int64','float64']).columns.tolist() 
```

```python
#Correlation HeatMap 
plt.figure(figsize=(8,6))
corr = df[numerical_features].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation  Between Numeric Features')
plt.show()
```

##### 0.0 to 0.2: Very Weak or No correlation<br>

```python
#scatterplot of hours studied per day and exam score
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='study_hours_per_day', y='exam_score', hue='part_time_job')
plt.title("Hours Studied (per day) vs Performance")
plt.show()
```

#### There is a strong relationship between the number of hours students studied per day and their corresponding exam scores. There is a clear linear relationship that can be observed in the scatterplot.The nature of the relationship is positive.<br>
#### There's a clear upward trend - as study hours increase, exam scores tend to increase.<br>
#### This pattern indicates that study time has a strong influence on performance.<br>
#### At around 6+ study hours/day, many students reach near 100% scores, suggesting that, beyond a certain point, more studying doesn't drastically boost scores further.<br>
#### Students with part-time jobs are generally well distributed and perform similarly to those without part-time jobs. This suggests that having a part-time job doesn't necessarily harm performance, especially if study time is maintained.<br>
#### Students who study <2 hours per day tend to have more varied and often lower scores - this indicates that low study hours negatively affect academic performance/ exam scores.

```python
# Scatterplot of exercise frequency and exam score
plt. figure(figsize=(8, 6))
sns. scatterplot(data=df, x='exercise_frequency', y='exam_score', hue='part_time_job')
plt. title("Exercise Frequency vs Performance")
plt. show( )
```

```python
# Boxplot of mental health rating and exam score
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='mental_health_rating', y='exam_score')
plt.title("Performance by Mental Health Score")
plt.show()
```

#### There seems to be a slight upward trend: students with higher mental health ratings (closer to 10) tend to have higher median exam scores.<br>
#### There are some outliers especially at lower mental health ratings (like 2), suggesting some students still perform very well even if their mental health is rated lower.<br>
#### Ratings from 7 to 10 seem to have slightly higher median scores compared to lower mental health ratings.<br>
#### Better mental health might be associated with slightly better academic performance, but there's still a lot of variability.<br>
